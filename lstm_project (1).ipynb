```python
faqs = """The sun was setting behind the mountains, casting a warm glow over the valley below. Birds chirped their evening songs as the day slowly came to an end. Sarah sat on the porch of her cabin, sipping a cup of tea and watching the scene unfold before her.

She had come to the mountains to escape the hustle and bustle of city life, seeking solace in nature's embrace. And now, as she gazed out at the peaceful landscape, she felt a sense of calm wash over her.

But her tranquility was soon interrupted by the sound of footsteps approaching from behind. She turned to see her neighbor, Mr. Thompson, making his way up the path.

"Evening, Sarah," he said with a smile as he reached the porch.

"Evening, Mr. Thompson," she replied, returning his smile. "Lovely night, isn't it?"

"It sure is," he agreed, taking a seat beside her. "Just can't beat the beauty of these mountains."

They sat in comfortable silence for a while, enjoying each other's company and the serenity of the evening. But eventually, Mr. Thompson broke the silence with a question.

"Say, Sarah, have you heard about the old mine up on the ridge?" he asked, his voice low and conspiratorial.

Sarah frowned slightly, feeling a sense of unease creep over her. The old mine was said to be haunted, its tunnels filled with the ghosts of miners long gone.

"I've heard rumors," she admitted cautiously. "But I've never been up there myself."

Mr. Thompson nodded knowingly. "Well, I was thinking of heading up there tomorrow morning to do some exploring. Thought you might like to join me."

Sarah hesitated, torn between her curiosity and her fear. She had always been intrigued by the stories surrounding the old mine, but she wasn't sure she wanted to risk venturing into its depths.

"I don't know, Mr. Thompson," she said slowly. "It sounds awfully dangerous."

He waved away her concerns with a dismissive gesture. "Nonsense! It'll be perfectly safe, I assure you. Besides, think of the adventure!"

Sarah chewed her lip nervously, weighing her options. In the end, her curiosity won out.

"Alright," she said, finally. "I'll go with you."

Mr. Thompson grinned triumphantly. "Excellent! We'll meet here at sunrise, then. Don't forget to bring a flashlight!"

With that, he bid her goodnight and made his way back down the path, leaving Sarah alone once more. She sat on the porch for a while longer, staring out at the fading light and wondering what awaited her in the depths of the old mine.

As the stars began to twinkle overhead, she finished her tea and retired to bed, her mind abuzz with thoughts of adventure and mystery."""
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=250)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(250, 100, input_length=47))
model.add(LSTM(150))
model.add(Dense(250, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=35)
```

```python
import numpy as np
```

```python
import time
text = "The"

for i in range(15):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=47, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
